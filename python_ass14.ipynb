```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

# Given data
months = np.arange(1, 13)
max_temp = np.array([39, 41, 43, 47, 49, 51, 45, 38, 37, 29, 27, 25])
min_temp = np.array([21, 23, 27, 28, 32, 35, 31, 28, 21, 19, 17, 18])

# Define a periodic function (for demonstration, you can choose the appropriate periodic function)
def periodic_func(x, a, b, c, d):
    return a * np.sin(b * x + c) + d

# Fit the data to the periodic function
popt_max, pcov_max = curve_fit(periodic_func, months, max_temp)
popt_min, pcov_min = curve_fit(periodic_func, months, min_temp)

# Plot the fit
plt.figure(figsize=(10, 6))
plt.plot(months, max_temp, 'ro', label='Max Temp Data')
plt.plot(months, periodic_func(months, *popt_max), 'r-', label='Max Temp Fit')
plt.plot(months, min_temp, 'bo', label='Min Temp Data')
plt.plot(months, periodic_func(months, *popt_min), 'b-', label='Min Temp Fit')
plt.xlabel('Months')
plt.ylabel('Temperature')
plt.title('Fitting Temperature Data to Periodic Function')
plt.legend()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load titanic data
url = 'https://raw.githubusercontent.com/Geoyi/Cleaning-Titanic-Data/master/titanic_original.csv'
titanic = pd.read_csv(url)

# Task 1: Create a pie chart presenting the male/female proportion
gender_proportion = titanic['sex'].value_counts()
plt.figure(figsize=(6, 6))
plt.pie(gender_proportion, labels=gender_proportion.index, autopct='%1.1f%%', startangle=90)
plt.title('Male/Female Proportion')
plt.show()

# Task 2: Create a scatterplot with Fare paid and Age, differing the plot color by gender
plt.figure(figsize=(10, 6))
colors = {'male': 'blue', 'female': 'red'}
plt.scatter(titanic['age'], titanic['fare'], c=titanic['sex'].map(colors))
plt.xlabel('Age')
plt.ylabel('Fare')
plt.title('Scatterplot of Fare vs Age (Color by Gender)')
plt.show()
```
