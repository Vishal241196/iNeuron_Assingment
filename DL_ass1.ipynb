Objective: - Predict Next Sequence
To start with deep learning, the very basic project that you can build is to
predict the next digit in a sequence.

Dataset: - Create a sequence like a list of odd numbers and then build a model
and train it to predict the next digit in the sequence.

Task: - A simple neural network with 2 layers would be sufficient to build the
model.

```python
import numpy as np
from tensorflow.keras.layers import Dense
from tensorflow.keras.models import Sequential
```

```python
# Step 1: Generate the dataset
odd_numbers = np.arange(1, 100, 2)  # Generate odd numbers from 1 to 99
```

```python
# Step 2: Preprocess the data
X = odd_numbers[:-1]  # Input sequence (all numbers except the last one)
y = odd_numbers[1:]   # Output sequence (all numbers except the first one)

# Step 3: Build the neural network model
model = Sequential([
    Dense(10, input_dim=1, activation='relu'),  # Input layer with 10 neurons and ReLU activation
    Dense(1)                                     # Output layer with 1 neuron (next number prediction)
])

# Step 4: Compile the model
model.compile(loss='mean_squared_error', optimizer='adam')

# Step 5: Train the model
model.fit(X, y, epochs=100, batch_size=1, verbose=2)

# Step 6: Predict the next number in the sequence
next_number = model.predict(np.array([99]))[0][0]
print("Predicted next number in the sequence:", int(round(next_number)))
```
